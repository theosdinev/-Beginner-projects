```python
import turtle

windo = turtle.Screen()
windo.title("Pong by Teo")
windo.bgcolor("black")
# 00 is at the center, left= - 400, right=400, up=300, down=-300
windo.setup(width=800, height=600)
windo.tracer(0)

# Paddle player A block
paddle_a = turtle.Turtle()
paddle_a.speed(0)   
paddle_a.shape("square")  #moje da e drugo
paddle_a.color("white")  #moje da e drug
paddle_a.shapesize(stretch_wid=5, stretch_len=1)
paddle_a.penup()
paddle_a.goto(-350, 0)  #koordinati

# Paddle player B block
paddle_b = turtle.Turtle()
paddle_b.speed(0)   
paddle_b.shape("square")  #moje da e drugo
paddle_b.color("white")  #moje da e drug
paddle_b.shapesize(stretch_wid=5, stretch_len=1)
paddle_b.penup()
paddle_b.goto(350, 0)  #koordinati

# Score
score_a = 0
score_b = 0


# Ball properties
ball = turtle.Turtle()
ball.speed(0)   
ball.shape("circle")  #moje da e drugo
ball.color("green")  #moje da e drug
ball.penup()
ball.goto(0, 0)  #koordinati
ball.dx = 0.4
ball.dy = -0.4


# Scoreboard block
pen = turtle.Turtle()
pen.speed(0)
pen.color("white")
pen.penup()
pen.hideturtle()
pen.goto(0,260)
pen.write("Player A: 0  ||  Player B :0", align="center", font=("Arial", 20, "normal"))

# Paddle Functions
def paddle_a_up():
    y = paddle_a.ycor()
    y += 20
    paddle_a.sety(y)
    
    
def paddle_a_down():
    y = paddle_a.ycor()
    y -= 20
    paddle_a.sety(y) 
    
    
def paddle_b_up():
    y = paddle_b.ycor()
    y += 20
    paddle_b.sety(y)
   
    
def paddle_b_down():
    y = paddle_b.ycor()
    y -= 20
    paddle_b.sety(y)  

    # Keyboard bindings
windo.listen()
windo.onkeypress(paddle_a_up, "w")
windo.onkeypress(paddle_a_down, "s")
windo.onkeypress(paddle_b_up, "Up")
windo.onkeypress(paddle_b_down, "Down")

while True:
    windo.update()
    
    # Ball movement
    ball.setx(ball.xcor() + ball.dx)
    ball.sety(ball.ycor() + ball.dy)
    
    
    # Border patrol:)
    if ball.ycor() > 290:
        ball.sety(290)
        ball.dy *= -1
        
        
    if ball.ycor() < -290:
        ball.sety(-290)
        ball.dy *= -1
    
    if ball.xcor() > 390:
        ball.goto(0, 0)
        ball.dx*= -1
        score_a += 1
        pen.clear()
        pen.write("Player A: {}  ||  Player B : {}".format(score_a, score_b), align="center", font=("Arial", 20, "normal"))
    
    if ball.xcor() < -390:
        ball.goto(0, 0)
        ball.dx*= -1
        score_b += 1
        pen.clear()
        pen.write("Player A: {}  ||  Player B : {}".format(score_a, score_b), align="center", font=("Arial", 20, "normal"))
    
    # Paddle and ball collision
    
    if (ball.xcor() > 340 and ball.xcor() < 350) and (ball.ycor() < paddle_b.ycor() + 40 and ball.ycor() > paddle_b.ycor() -40):
        ball.setx(340)
        ball.dx *= -1
    
    if (ball.xcor() < -340 and ball.xcor() > -350) and (ball.ycor() < paddle_a.ycor() + 40 and ball.ycor() > paddle_a.ycor() -40):
        ball.setx(-340)
        ball.dx *= -1
        
        
    
    
```
